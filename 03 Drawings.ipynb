### Rectangle using matplotlib

```python
x_min,y_min,x_max,y_max = 1,2,4,5

#creat a rectangle patch

rectangle = plt.Rectangle((x_min,y_min), x_max - x_min , y_max-y_min , color = 'cyan',alpha = 0.5)

# add the rectangle to plot and costomize label

plt.plot([x_min,x_max],[y_min,y_min],'r-',label = 'X-axis')
plt.plot([x_min,x_min],[y_min,y_max],'r-',label = 'Y-axis')
plt.gca().add_patch(rectangle)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Rectangle')
plt.legend()

# show rectangle

plt.show()
```

### Ellipse using matplotlib

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
```

```python
# define all parameters of ellipse
center = (0,2)
width = 8
height = 5
angle = 90

# create ellipse object
ellipse = Ellipse(xy = center,width=width, height=height,angle=angle,color = 'r')

# add ellipse to the plot
fig,ax = plt.subplots()
ax.add_patch(ellipse)
ax.set_aspect('equal')        # maintain equal aspect ratio for correct shape

# set limits slightly wider than ellipse
ax.set_xlim(-width-width / 5 , width+width/5)
ax.set_ylim(-height-height/5,height+height/5)

# Add label and Title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Ellipse Plot')

# show the plot
plt.show()
```

### GUI Design

```python
# Create the main window
root = Tk()
root.geometry('1000x800')
root.title('Drawing Canvas')

# Create the canvas widget
canvas = Canvas(root, width = 1000, height = 800, bg  ='white')

canvas.pack()

# Define drawing functions
def draw_line(event):
    x1, y1 = event.x, event.y
    canvas.create_line(x1, y1, x1 + 2, y1 + 2, width = 2, fill = 'black')
    
def draw_circle(event):
    x1, y1 = event.x, event.y
    canvas.create_oval(x1 - 5, y1 - 5, x1 + 5, y1 + 5, fill = 'yellow')
    
# Bind mouse events to drawing functions
canvas.bind("<Button-3>", draw_circle)
canvas.bind("<B1-Motion>", draw_line)

# Start the main loop
root.mainloop()
```

### Mouse Operations

```python
import cv2
```

```python
img = cv2.imread('lenna.png')

def draw_circle(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),10,(1,1,1),-1)
        
cv2.namedWindow('image')
cv2.setMouseCallback('image',draw_circle)

while True:
    cv2.imshow('image',img)
    if cv2.waitKey(1) == ord('q'):
        break
        
cv2.destroyAllWindows()
```

### Image Annotation

```python
# Define text and rectangle properties
text = "Sample"
font_face = cv2.FONT_HERSHEY_COMPLEX
font_scale = 1.0
color = (0, 255, 145)    # Green
thickness = 2

# Calculate text width and height
text_size, baseline = cv2.getTextSize(text, font_face, font_scale, thickness)

# Define text and rectangle positions
text_origin = (200,200)
box_origin = (text_origin[0] - 10, text_origin[1] - baseline - 20)
box_width, box_height = text_size[0] + 20, text_size[1] + 20

image = cv2.imread("lenna.png")

# Draw the rectangle and text
cv2.rectangle(image, box_origin, (box_origin[0] + box_width,
                                 box_origin[1] + box_height), color, thickness)
cv2.putText(image, text, text_origin, font_face, font_scale, color, thickness)

# Display the annotated image
cv2.imshow("Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
