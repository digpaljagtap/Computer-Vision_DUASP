### Image Translation / shifting

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('lenna.png')

# Store height and weight of the image
height, width = image.shape[:2]

quarter_height, quarter_width = height / 6, width / 12

T = np.float32([[1, 0, quarter_width], [0, 1, quarter_height]])

# We use warpAffine to transform the image using the matrix, T
img_translation = cv2.warpAffine(image, T, (width, height))

cv2.imshow("Originalimage", image)
cv2.imshow('Translation', img_translation)
cv2.waitKey()

cv2.destroyAllWindows()
```

### Image Rotation

```python
# load image
image = cv2.imread('lenna.png')
 
#rotate image clockwise
rotated_image = cv2.rotate(image,cv2.ROTATE_180)

# show rotated image
cv2.imshow('rotated_image',rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Using scikit.image

```python
from skimage.transform import resize
```

```python
image = cv2.imread('lenna.png')

resized_image = resize(image,(1000,1200),anti_aliasing=True)

resized_image = resize(image,(1300,1300),anti_aliasing=True, order =3)

cv2.imshow('resized_image',resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread('lenna.png')

flipped_horizontal = cv2.flip(image,0)

flipped_vertical = cv2.flip(image,1)

flipped_diagonal = cv2.flip(image,-1)

cv2.imshow('Original',image)
cv2.imshow('Horizontal',flipped_horizontal)
cv2.imshow('Vertical',flipped_vertical)
cv2.imshow('Diagonal',flipped_diagonal)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Image Flipping using PIL

```python
# load image
image = Image.open('lenna.png')
# flip image horizontal
flipped_image = image.transpose(Image.FLIP_LEFT_RIGHT)
flipped_image.show()
# flip vertical
flipped_image = image.transpose(Image.FLIP_TOP_BOTTOM)
flipped_image.show()
# flip both axis(diagonal flip)
flipped_image = image.transpose(Image.TRANSPOSE)
flipped_image.show()
```
