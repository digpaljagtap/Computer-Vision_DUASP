Thresholding
--
It is a fundamental technique in CV used to segment an image into foreground and background objects<br>
It works by classifying each pixel in the image as either belonging to the object of interest or the background based on their intensity value

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image1 = cv2.imread('lenna.png')

# it is applied over the image input with applied parameters
# to convert image in grayscale
img = cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)

# applying diff thresholding technique on the input image 
# all pixels value above 120 will be set to 255
ret, thresh1 = cv2.threshold(img,120,255,cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img,120,255,cv2.THRESH_BINARY_INV)

plt.figure(figsize=(15,10))
plt.subplot(121)
plt.title('Binary Thresholding')
plt.imshow(thresh1,cmap='gray')
plt.subplot(122)
plt.title('Inverse Binary Thresholding')
plt.imshow(thresh2,cmap='gray')
```

```python
image1 = cv2.imread('lenna.png')

# it is applied over the image input with applied parameters
# to convert image in grayscale
img = cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)

# applying diff thresholding technique on the input image 
# all pixels value above 120 will be set to 255
ret, thresh1 = cv2.threshold(img,120,255,cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img,120,255,cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(img,120,255,cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(img,120,255,cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(img,120,255,cv2.THRESH_TOZERO_INV)


plt.figure(figsize=(15,10))
plt.subplot(151)
plt.title('Binary Thresholding')
plt.imshow(thresh1,cmap='gray')
plt.subplot(152)
plt.title('Inverse Binary Thresholding')
plt.imshow(thresh2,cmap='gray')
plt.subplot(153)
plt.title('Trunc Thresholding')
plt.imshow(thresh3,cmap='gray')
plt.subplot(154)
plt.title('Tozero Thresholding')
plt.imshow(thresh4,cmap='gray')
plt.subplot(155)
plt.title('Inverse Tozero Thresholding')
plt.imshow(thresh5,cmap='gray')
```

Otsu Thresholding
--
It is popular and effective automatic thresholding technique used in image processing to seperate an image into foreground and background based on pixel intensity<br>
it aims to find the threshold value that maximize the between class variance , essentially creating the best possible separation between the two classes 

```python
img = cv2.imread('lenna.png')

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

_, thresh_otsu = cv2.threshold(gray,120,255,cv2.THRESH_OTSU)

cv2.imshow('Original Image', img)
cv2.imshow('Otsu Thresholding', thresh_otsu)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from skimage import io, filters
image = io.imread('lenna.png', as_gray = True)
```

```python
threshold = filters.threshold_otsu(image)
binary_image = image > threshold
```
